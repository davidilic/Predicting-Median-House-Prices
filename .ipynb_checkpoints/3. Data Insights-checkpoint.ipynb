### 3.2 VISUALISATIONS

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = exploration_set['longitude']
y = exploration_set['latitude']
z = exploration_set['median_house_value']
plt.figure(figsize=(20,12))
geo_fig = plt.scatter(x, y, alpha=0.2, linewidths=1, edgecolor='black', 
            s = exploration_set['population']/30, label='population',
            c=z, cmap=plt.get_cmap("jet")
)
plt.colorbar(geo_fig)
plt.legend();
```

* Distance to coastline seems to be a big factor.
* The ocean_proximity attribute in the dataset is a discrete value coded as a python object. 
* The discrete ocean_proximity variable loses lots of explanatory power. Should be turned into a continuous variable.
* The ocean_proximity could be turned continous using linear algebra, latitude, longitude and a coastline approximation.

### 3.4 ATTRIBUTE COMPOSITION

```python
exploration_set.info()
```

* Possible successfull compositions:
    * rooms_per_house = total_rooms / households
    * bedrooms_per_house = total_bedrooms / households
    * population_per_house = population / households
    * population_per_room = population / total_rooms
    * bedrooms_per_room = total_bedrooms / total_rooms

```python
exploration_set['rooms_per_house'] = exploration_set['total_rooms'] / exploration_set['households']
exploration_set['bedrooms_per_house'] = exploration_set['total_bedrooms'] / exploration_set['households']
exploration_set['population_per_house'] = exploration_set['population'] / exploration_set['households']
exploration_set['population_per_room'] = exploration_set['population'] / exploration_set['total_rooms']
exploration_set['bedrooms_per_room'] = exploration_set['total_bedrooms'] / exploration_set['total_rooms']

exploration_set.corr()['median_house_value'].sort_values()
```

* Successfull compositions:
    * rooms_per_house
    * bedrooms_per_room
